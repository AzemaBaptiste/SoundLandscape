```python
import pandas as pd
import numpy as np
import json

def jsonToCSV():
    start = 0
    end = 1000
    df = pd.DataFrame()
    while end != 5000:
        songPlaylistArray = []
        path = '../data/raw/music/mpd.slice.' + str(start) + "-" + str(end-1) + '.json'
        d = json.load(open(path, 'r'))
        thisSlice = pd.DataFrame.from_dict(d['playlists'], orient='columns')
        for index, row in thisSlice.iterrows():
            for track in row['tracks']:
                
                songPlaylistArray.append([track["artist_uri"], track["duration_ms"], track['track_uri'], track['artist_name'], track['track_name'], row['pid'], row["name"]])
        songPlaylist = pd.DataFrame(songPlaylistArray, columns=["artist_uri", "duration_ms", 'trackid', 'artist_name', 'track_name', 'pid', "name"])
        df = pd.concat([df, songPlaylist])
        # songPlaylist.to_csv('data_CSV/' + str(start) + "-" + str(end-1) + '.csv', index=False)
        start += 1000
        end += 1000
    
    return df
```

```python
df = jsonToCSV()
```
